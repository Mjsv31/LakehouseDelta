## Funciones

```python
# - Funcion que realiza la carga inicial de archivos parquet
# - en la capa landing
def fGeneraParquet(pArchivo):

    # -- Procesa archivos de minio
    vPath = f'{vData}/{pArchivo}'

    # -- Mensaje a pantalla
    print("")
    print(f'Ingestando archivo {pArchivo}')
    print("")

    # -- Genera dataframe a partir del archivo descargado
    df_archivo = spark.read.parquet(vPath)
    
    # -- Escribe archivo parquet en capa landing
    df_archivo.write.format("parquet").mode("overwrite").save(f"{vPathLand}{pArchivo}")
```

## Ingesta archivos parquet

```python
#  -- Lista de archivos a cargar
vArchivos = [
    'customers.parquet',
    'dealerships.parquet',
    'employees.parquet',
    'payments.parquet',
    'products.parquet',
    'products_costs.parquet',
    'promotions.parquet',
    'salaries.parquet',
    'sales_transactions.parquet',
    'dim_dates.parquet'
    ]

# Por cada archivo invoca a la funcion fGeneraParquet
for Archivo in vArchivos:
    fGeneraParquet(Archivo)
```
